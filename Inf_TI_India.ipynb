## 3. Desarrollo de objetivos

```python
# Rating más alto de las empresas en la India
rmax = chipo ['Rating']. max ()
print('El Rating mas alto es: ',+rmax)
```

```python
# Cantidad de Compañías Contratantes en la India

empresa = chipo ['Company Name']
empresa = empresa.drop_duplicates().value_counts()
empresa
print("\nSegun los datos analizados, la cantidad de empresas contrtantes en la India son: ",+empresa.sum(),'\nY las empresas, son las siguientes:\n')
print(empresa)
```

```python
# Cantidad de Puesto de trabajo entre las empresas de la India

puesto = chipo ['Job Title']
puesto = puesto.drop_duplicates().value_counts()
puesto
print("\nSegun los datos analizados, la cantidad de puestos de trabajo en la India son: ",+puesto.sum(),'\nY los puestos de trabajo, son los siguientes:\n')
print(puesto)
```

```python
#Salario más alto en los que oscilan los puesto de trabajo en la India

salario = chipo ['Salary']
salario = salario.drop_duplicates().value_counts()

smax = chipo ['Salary']. max ()

print('Segun los datos analizados, los salarios Oscilan entre:\n')
print(salario)
print('\nEl salario mas alto es: ',+smax)
```

```python
#Cantidad de desarrolladores contratados por las empresas en la India

desarrolladores = chipo ['Salaries Reported']
desarrolladores = desarrolladores.sum()
print("\nSegun los datos analizados, la cantidad de desarrolladores contratados en la India son ",+desarrolladores.sum(),' dearrolladores')
```

```python
#cantidad de empresas por localidad en la India

localidad = chipo.groupby (['Location'])['Salaries Reported'].sum()
emp = chipo ['Salaries Reported'].sum()

print("Segun los datos analizados, la cantidad total de empresas de desarrollo en la india es :",+emp.sum(),"\n\nla cantidad de empresas por localidad el la inddia es:")
print(localidad)
```

```python
#Tipo de contratacion que se implementa en las empresas de Desarrollo en la India
tipocont = chipo ['Employment Status']
tipocont = tipocont.drop_duplicates().value_counts()
tipocont
print("\nSegun los datos analizados, el tipo de contratacion que se implementa en la India son: \n")
print(tipocont)
```

```python
#Rol a desempeñar en las empresas de la India

rol = chipo ['Job Roles']
rol = rol.drop_duplicates().value_counts()
rol
print("\nSegun los datos analizados, los roles que se desempeña en las empresas en la India son: \n")
print(rol)
```

## 4. Gráficos

- Roles y tipos de contrato.

```python
# Cantidad de Compañías Contratantes en la India
# locaciones donde se encuentran las empresas

tipocont = chipo ['Employment Status']
tipocont = tipocont.drop_duplicates().value_counts().sum()

roles = chipo ['Job Roles']
roles = roles.drop_duplicates().value_counts().sum()

# put them into a list called proportions
proportions = [roles, tipocont]

# Create a pie chart
plt.pie(
    # using proportions
    proportions,
    
    # with the labels being officer names
    labels = ['Roles', 'Tipo de Contrato'],
    
    # with no shadows
    shadow = True,
    
    # with colors
    colors = ['yellow','orange'],
    
    # with one slide exploded out
    explode = (0.15 , 0.5),
    
    # with the start angle at 90%
    startangle = 100,
    
    # with the percent listed as a fraction
    autopct = '%1.1f%%'
    )

# View the plot drop above
plt.axis('equal')

# Set labels
plt.title("EMPRESAS DE DESARROLLO DE LA INDIA")

# View the plot
plt.tight_layout()
plt.show()

print('Con esta grafica se demuestra que para cada rol, no existe un tipo de contrato, por lo tanto,\nvarios roles, estan agrupados en un tipo de contrato.')
```
